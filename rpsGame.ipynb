```python
import random, sys
```

```python
print("Are you up for Rock Paper Scissors challenge?")
# following variables keep track of the stats
wins = 0
losses = 0
ties = 0
while True: # main game loop
    print('%s Wins, %s Losses, %s Ties'%(wins,losses,ties))
    while True: # player move loop
        print("Enter your move: (r)ock (p)aper (s)cissors or (q)uit")
        playerMove = input()
        if playerMove == "q":
            sys.exit() # quit the game/program
        if playerMove =='r' or playerMove =='p' or playerMove =='s':
            break # since expected input, exit input loop
        print('Please enter r, p, s, or q to proceed.')
    # player loop complete, now what to display, before selecting computer move
    if playerMove == 'r':
        print('Rock vs. ')
    elif playerMove =='p':
        print('Paper vs. ')
    elif playerMove =='s':
        print('Scissors vs. ')
    # Computer move
    computerMove = random.randint(1,3)
    if computerMove ==1:
        computerMove = 'r'
        print('Rock')
    elif computerMove ==2:
        computerMove = 'p'
        print('Paper')
    elif computerMove == 3:
        computerMove = 's'
        print('Scissors')
    # Now that computer's move has been recorded, display the score
    if playerMove == computerMove:
        print("""It's a tie!""")
        ties += 1
    elif playerMove == 'r' and computerMove == 'p':
        print("You lose!")
        losses += 1
    elif playerMove == 'p' and computerMove == 's':
        print("You lose!")
        losses += 1
    elif playerMove == 's' and computerMove == 'r':
        print("You lose!")
        losses += 1
    elif playerMove == 'r' and computerMove == 's':
        print("You win!")
        wins += 1
    elif playerMove == 'p' and computerMove == 'r':
        print("You win!")
        wins += 1
    elif playerMove == 's' and computerMove == "p":
        print("You win!")
        wins += 1
```
